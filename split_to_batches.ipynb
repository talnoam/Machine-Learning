```python
def build_block(jet):                      #this function makes a nxnx6 matrix
    block = []
    nodes_num = jet.shape[0]
    graph = np.empty((nodes_num, nodes_num, 6))
    index = 0
    for i in range(6):
        graph[:, :, i] = np.diag(jet[:, i])
        i == index
        block.append(graph[:, :, i])
        i += 1 
    block = np.array(block)
    return block
```

```python
# organizes all the blocks in one place
index = 0
blocks = []
for i in range(18865):
    i == index
    jet = ds.__getitem__(i)['node_features']
    jet_block = build_block(jet)
    blocks.append(jet_block)
    i += 1
```

```python
def same_size(size):
    one_size = []
    for i in range(len(blocks)):
        if blocks[i].shape[1] == size:
            one_size.append(blocks[i])
    return one_size
```

```python
def group_same_size(blocks):
    sizes = list(map(lambda t: t.shape[1], blocks))
    r_blocks = []
    index = 0
    for size in range(len(sizes)):
        size == index
        one_size = same_size(size)
        if one_size != []:
            r_blocks.append(one_size)
        one_size = []
        size += 1
    return r_blocks
```

```python
block_size_list = group_same_size(blocks)
```
